```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

from tensorflow.keras.datasets import mnist
```

```python
seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits
```

## EX 12.1

We train a DNN by increasing the number of epochs to an adequate value. We try other optimizers than SGD and show accuracy and loss for training and validation data.

```python
# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
print('X_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print()
```

```python
# cast floats to single precision
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255
```

```python
# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
```

```python
def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    return model
```

```python
def compile_model(optimizer):
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy, optimizer=optimizer, metrics=['acc'])
    return model
```

```python
# training parameters
batch_size = 32
#raised the number of epochs from 5 to 10
epochs = 10
```

```python
##SGD###

model_DNN = compile_model(SGD())
# train DNN and store training info in history
history = model_DNN.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, verbose=0, validation_data=(X_test, Y_test))
# evaluate model
fig = plt.figure(figsize=(14, 4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
# summarize history for accuracy
ax1.plot(history.history['acc'])
ax1.plot(history.history['val_acc'])
ax1.set_title('accuracy -> sgd optimizer')
ax1.set(xlabel='epoch', ylabel='model accuracy')
ax1.legend(['train', 'test'], loc='best')
ax1.grid(True)
# summarize history for loss
ax2.plot(history.history['loss'])
ax2.plot(history.history['val_loss'])
ax2.set_title('loss -> sgd optimizer')
ax2.set(xlabel='epoch', ylabel='model loss')
ax2.legend(['train', 'test'], loc='best')
ax2.grid(True)
plt.show()
    
predictions = model_DNN.predict(X_test)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)
plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

## EX 12.2

Now we change the architecture from DNN to CNN and evaluate the performance exactly as before. We now use only one optimizer because the simulation is time 
expensive for 10 epochs.

```python
from tensorflow.keras.layers import Flatten, Conv2D, MaxPooling2D
```

```python
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols) / 255
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols) / 255
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1) / 255
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1) / 255
    input_shape = (img_rows, img_cols, 1)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
# Express labels in categorical notation
Y_train = keras.utils.to_categorical(Y_train)
Y_test = keras.utils.to_categorical(Y_test)
```

```python
def create_CNN():
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 10 filters (dimensionality of output space)
    model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Dropout(0.2))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    model.add(Dropout(0.2))
    model.add(Flatten())
    model.add(Dense(100, activation='relu'))
    model.add(Dense(10, activation='softmax'))
    model.compile(loss=keras.losses.categorical_crossentropy, optimizer=SGD(), metrics=['acc'])
    return model
```

```python
# training parameters
batch_size = 32
epochs = 10

# create the deep conv net
model_CNN=create_CNN()

# train CNN
history = model_CNN.fit(X_train, Y_train, batch_size=batch_size, epochs=epochs, verbose=1, validation_data=(X_test, Y_test))

# evaliate model
score = model_CNN.evaluate(X_test, Y_test, verbose=0)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
#SGD CNN
fig = plt.figure(figsize=(14, 4))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
# summarize history for accuracy
ax1.plot(history.history['acc'])
ax1.plot(history.history['val_acc'])
ax1.set_title('CNN accuracy -> sgd optimizer')
ax1.set(xlabel='epoch', ylabel='model accuracy')
ax1.legend(['train', 'test'], loc='best')
ax1.grid(True)
# summarize history for loss
ax2.plot(history.history['loss'])
ax2.plot(history.history['val_loss'])
ax2.set_title('CNN loss -> sgd optimizer')
ax2.set(xlabel='epoch', ylabel='model loss')
ax2.legend(['train', 'test'], loc='best')
ax2.grid(True)
plt.show()
```

## EX 12.3

We use our CNN to recognize 10 handwritten digits.  
We use PIL/Pillow library to import digits created with gimp.

```python
from PIL import Image
import os
```

```python
fig, ax = plt.subplots(2, 5,figsize = (15,5))
ax = ax.flatten()
digit = [str(d) for d in range(10)]
for d_c, dig in zip(range(10), digit):
    digit_filename = "digit/"+str(dig)+".png"
    digit_in = Image.open(digit_filename).convert('L')
    ydim, xdim = digit_in.size
    pix=digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for i in range(xdim):
            data[i,j]=pix[j,i]    
    ax[d_c].imshow(data, cmap = "gray")
    if keras.backend.image_data_format() == "channels_first":
        data = data.reshape(1, 1, 28, 28)
    else:
        data = data.reshape(1, 28, 28, 1)
    data /= 255
    ax[d_c].axis("off")
    ax[d_c].set_title(f"Digit predicted: {np.argmax(model_CNN.predict(data))}")
fig.tight_layout()
plt.show()
    
```
