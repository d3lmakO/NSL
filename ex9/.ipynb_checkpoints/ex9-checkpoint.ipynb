```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

## EX 9.1

We use a Genetic Algorithm (GA) to solve the Travelling salesman problem (TSP).  
The TSP is solved for cities arranged on a **circumference** and inside a **square**.

```python
with open("data/N_data_9_1.txt") as file:
    N_cities = int(file.readline())
    N_paths = int(file.readline())
    N_generations = int(file.readline())
    p_selection = float(file.readline())
    p_crossover = float(file.readline())
    p_mutation = float(file.readline())
print(f"Number of cities = {N_cities}")
print(f"Number of paths = {N_paths}")
print(f"Number of generations = {N_generations}")
print(f"Probability of selection = {p_selection}")
print(f"Probability of crossover = {p_crossover}")
print(f"Probability of mutation = {p_mutation}")
```

### 9.1.1 

Circumference

```python
cities_circle = np.loadtxt("data/9_1_circle_cities.txt")
circle_paths = np.loadtxt("data/9_1_circle_paths_out.txt")
best_c_path = np.loadtxt("data/9_1_circle_best_path.txt").astype(int)
```

```python
x_c = np.arange(circle_paths[:,0].size)
best_circle = circle_paths[:,0]
ave_best_half_circle = circle_paths[:,1]
plt.figure(figsize=(9, 7))
plt.grid(True)
plt.plot(x_c, best_circle, label="Best circumference path");
plt.plot(x_c, ave_best_half_circle, label="Average best half population");
plt.xlim(0, N_generations)
plt.legend()
plt.title("Path cost ($L^2$)")
plt.xlabel("Generation")
plt.ylabel("Cost")
plt.show()
```

**Best path in cartesian coordinates**

```python
idx_c = np.append(best_c_path, best_c_path[0])

plt.figure(figsize=(8, 8))
plt.plot(cities_circle[idx_c, 1], cities_circle[idx_c, 2], '-o');
plt.grid()
plt.title("Path on circumference")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

### 9.1.2

Square

```python
cities_square = np.loadtxt("data/9_1_square_cities.txt")
square_paths = np.loadtxt("data/9_1_square_paths_out.txt")
best_s_path = np.loadtxt("data/9_1_square_best_path.txt").astype(int)
```

```python
x_s = np.arange(square_paths[:,0].size)
best_square = square_paths[:,0]
ave_best_half_square = square_paths[:,1]
plt.figure(figsize=(9, 7))
plt.grid(True)
plt.plot(x_s, best_square, label="Best square path");
plt.plot(x_s, ave_best_half_square, label="Average best half population");
plt.xlim(0, N_generations)
plt.legend()
plt.title("Path cost ($L^2$)")
plt.xlabel("Generation")
plt.ylabel("Cost")
plt.show()
```

**Best path in cartesian coordinates**

```python
idx_s = np.append(best_s_path, best_s_path[0])

plt.figure(figsize=(8, 8))
plt.plot(cities_square[idx_s, 1], cities_square[idx_s, 2], '-o');
plt.grid()
plt.title("Path in square")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```
