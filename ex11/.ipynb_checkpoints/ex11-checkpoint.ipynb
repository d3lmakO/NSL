## EX 11.1

We are considering the function:
$$
f(x)=2x+1
$$

We fit the function exploring the dependences of the fit from the number of epochs, $N_{\mathrm{epochs}}$, the number of data points $N_{\mathrm{train}}$ and on the noise $\sigma$.

```python
def one_layer_linear_fit(N_epochs, N_train, sigma):
    # target parameters of f(x) = m*x + b
    m = 2 # slope
    b = 1 # intersect
    # generate training inputs
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, N_train)
    x_valid = np.random.uniform(-1, 1, N_train // 10)
    x_valid.sort()
    y_target = m * x_valid + b # ideal (target) linear function
    y_train = np.random.normal(m * x_train + b, sigma) 
    y_valid = np.random.normal(m * x_valid + b, sigma)
    # plot validation and target dataset
    fig = plt.figure(figsize=(14, 4))
    # Adds subplot on position 1
    ax1 = fig.add_subplot(121)
    # Adds subplot on position 2
    ax2 = fig.add_subplot(122)
    ax1.plot(x_valid, y_target, color='b', label='target')
    ax1.scatter(x_train, y_train, color='r', label='data')   
    ax1.set(xlabel='x', ylabel='$f(x)$')
    #one neuron model
    model = tf.keras.Sequential()
    model.add(Dense(1, input_shape=(1,)))
    # compile the model choosing optimizer, loss and metrics objects
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    # get a summary of our composed model
    # fit model
    # 32 batch size
    # report training progress against validation data
    history = model.fit(x=x_train, y=y_train, 
              batch_size=32, epochs=N_epochs,
              shuffle=True, # a good idea is to shuffle input before at each epoch
              validation_data=(x_valid, y_valid), verbose=0)
    # return weights and biases
    print("f(x) =", model.get_weights()[0][0][0], "x +", model.get_weights()[1][0])
    # evaluate model
    score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    # evaluate model with the exact curve
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)
    print('Exact curve')
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    #predicted data
    x_predicted = np.random.uniform(-1, 1, 100)
    y_predicted = model.predict(x_predicted)
    ax1.scatter(x_predicted, y_predicted, color='g', label='predicted')
    ax1.legend(loc='best')
    ax1.grid(True)  
    # look into training history
    # Plot training & validation loss values
    ax2.plot(history.history['loss'], label='train')
    ax2.plot(history.history['val_loss'], label='test')
    ax2.set_title('Model loss')
    ax2.set(xlabel='Epoch', ylabel='Loss')
    ax2.legend(loc='best')
    ax2.grid(True)  
    plt.show()
```

## EX 11.2

We extend the neural model to obtain a fit of a polynomial of order three:  
$$
f(x)=4-3x-2x^2+3x^3
$$
for $x \in [-1,1]$.


We try good choices of:

- the number of layers
- the number of neurons in each layer
- the activation function
- the optimizer
- the loss function

```python
def multilayer_non_linear_fit(N_epochs, N_train, sigma, n_neurons, n_layers, p_optimizer, p_activation, p_loss, l_i, r_i):    
    # target parameters of f(x) = ax^3 + bx^2 + cx + d
    a = 3 
    b = -2
    c = -3
    d = 4
    f = lambda x: a*x*x*x + b*x*x + c*x + d
    # generate training inputs
    np.random.seed(0)
    x_train = np.random.uniform(-1, 1, N_train)
    x_valid = np.random.uniform(l_i, r_i, N_train // 10)
    x_valid.sort()
    y_target = f(x_valid)
    y_train = np.random.normal(f(x_train), sigma) 
    y_valid = np.random.normal(f(x_valid), sigma)
    # model
    model = tf.keras.Sequential()
    model.add(Dense(n_neurons, input_shape=(1,), activation=p_activation))
    # internal layer
    for i in range(1, n_layers - 1):
        model.add(keras.layers.Dense(n_neurons, activation=p_activation))
    # output layer must have only one output neuron
    model.add(Dense(1, activation=p_activation))
    model.compile(loss=p_loss, optimizer=p_optimizer, metrics=['mse'])
    history = model.fit(x=x_train, y=y_train, batch_size=32, epochs=N_epochs,
              shuffle=True, # a good idea is to shuffle input before at each epoch
              validation_data=(x_valid, y_valid), verbose=0)
    # return weights
    #print(model.get_weights())
    # evaluate model
    score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    # evaluate model with the exact curve
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)
    print('Exact curve')
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    # plot validation and target dataset
    fig = plt.figure(figsize=(14, 4))
    # Adds subplot on position 1
    ax1 = fig.add_subplot(121)
    # Adds subplot on position 2
    ax2 = fig.add_subplot(122)
    ax1.plot(x_valid, y_target, color='b', label='target')
    ax1.scatter(x_train, y_train, color='r', label='data')   
    ax1.set(xlabel='x', ylabel='$f(x)$')
    #predicted data
    x_predicted = np.random.uniform(l_i, r_i, N_train // 10)
    y_predicted = model.predict(x_predicted)
    ax1.scatter(x_predicted, y_predicted, color='g', label='predicted')
    ax1.legend(loc='best')
    ax1.grid(True)
    # Plot training & validation loss values
    ax2.plot(history.history['loss'], label='train')
    ax2.plot(history.history['val_loss'], label='test')
    ax2.set_title('Model loss')
    ax2.set(xlabel='Epoch', ylabel='Loss')
    ax2.legend(loc='best')
    ax2.grid(True)  
    plt.show()
```

## EX 11.3

We extend the model to fit a trigonometric 2D function:  
$$f(x,y) = \sin(x^2+y^2)$$  
in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
# generate training inputs
sigma = 1.0
np.random.seed(0)
x1_train = np.random.uniform(-3./2., 3./2., 10000)
x2_train = np.random.uniform(-3./2., 3./2., 10000)
x1_valid = np.random.uniform(-3./2., 3./2., 1000)
x2_valid = np.random.uniform(-3./2., 3./2., 1000)
x1_valid.sort()
x2_valid.sort() 
y_target = np.sin(x1_valid*x1_valid + x2_valid*x2_valid)
y_train = np.random.normal(np.sin(x1_train*x1_train + x2_train*x2_train), sigma)
y_valid = np.random.normal(np.sin(x1_valid*x1_valid + x2_valid*x2_valid), sigma)
model = keras.models.Sequential()
model.add(Dense(20, input_shape=(2,), activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1,))
model.compile(loss='mse', optimizer='adam', metrics=['mse'])
# EarlyStopping: Stop training when a monitored metric has stopped improving.
#check at end of every epoch whether the loss is no longer decreasing, 
 #considering the min_delta and patience if applicable. Once it's found no longer decreasing, 
#model.stop_training is marked True and the training terminates.
earlystop = tf.keras.callbacks.EarlyStopping(monitor='mse', min_delta=0.0001, patience=8, verbose=0, mode='auto')
train = np.transpose(np.array([x1_train, x2_train]))
valid = np.transpose(np.array([x1_valid, x2_valid]))
history = model.fit(x=train, y=y_train, batch_size=32, epochs=500, shuffle=True,
        callbacks = [earlystop],
        validation_data=(valid, y_valid), verbose=0)
```

```python
plt.figure(figsize=(11,4))
plt.grid(True)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.show()
```

```python
x1_predicted = np.random.uniform(-3./2., 3./2., 1000)
x2_predicted = np.random.uniform(-3./2., 3./2., 1000)
predicted = np.transpose(np.array([x1_predicted, x2_predicted]))
y_predicted = model.predict(predicted)

fig = plt.figure(figsize=(10,10))
ax =fig.add_subplot(projection='3d')
X = np.arange(-3./2., 3./2., 0.02)
Y = np.arange(-3./2., 3./2., 0.02)
X, Y = np.meshgrid(X, Y)
R = X**2 + Y**2
Z = np.sin(R)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('sin($x^2+y^2$)')
ax.dist = 12
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis, linewidth=0, antialiased=True)
ax.scatter(x1_predicted,x2_predicted, y_predicted,color='r', marker='o', s=1)
plt.show()
```
