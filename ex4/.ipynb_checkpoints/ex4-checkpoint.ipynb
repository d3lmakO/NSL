```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker
import math
```

## EX 4.2

We perform MD simulations via a Lennard-Jones model of **Argon** for three phases: 
- solid $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)
- liquid $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)
- gas $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

We show pictures of the progressive average values of:
- energy per particle $E/N$
- potential energy per particle $U/N$
- kinetic energy per particle $K/N$
- temperature $T$
- pressure $P$
in reduced units.

Before to measure this quantites we need a period of equilibration to stabilize the temperature, so after a certain number of steps $T$ will be near the desired temperature
$T^\star$

### 4.2.1  Solid phase ARGON

**EQUILIBRATION**

Starting T = 1.4

```python
n_equilibration = 10000

solid_e = np.loadtxt("equilibration/phase/solid_phase.txt")
e_kin_e = solid_e[:,0]
e_pot_e = solid_e[:,1]
e_tot_e = solid_e[:,2]
temp_e = solid_e[:,3]
x = np.arange(n_equilibration)

tick_spacing = 0.6

fig = plt.figure(figsize=(14, 4))
ax1 = fig.add_subplot(121)
ax1.plot(x, e_kin_e, label="$E_\mathrm{kin}$")
ax1.plot(x, e_pot_e, label="$E_\mathrm{pot}$")
ax1.plot(x, e_tot_e, label="$E_\mathrm{tot}$")
ax1.plot(x, temp_e, label="$T$")
ax1.set(xlabel="#steps", ylabel="$E$ -- $T$")
ax1.set_xlim(0, n_equilibration)
ax1.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax1.set_title('solid phase')
ax1.legend(loc='best')
ax1.grid(True)
plt.show()
```

Starting from a perfect crystalline configuration, during the first time steps of the simulation $T$ will not conserve its value.  
T is linked with the kinetic energy and the kinetic energy falls down. It happens because the system starts in a perfect crystalline configuration with very low potential
energy and the simulation creates disorder in the system and so the potential energy goes up. If the potential energy goes up then the kinetic energy goes down because
we are in the microcanonical ensamble and the total energy is conserved.

**MEASURE**

After equilibration, we start with a proper configuration to measure ours quantites.  
For equilibration we use less than 10000 steps, 5000 are more than appropriate, because the system equilibrate quickly.

```python
sx1, sy1, serr1 = np.loadtxt("data/phase/solid_output_ekin.dat", usecols=(0,2,3), unpack='true')
sx2, sy2, serr2 = np.loadtxt("data/phase/solid_output_epot.dat", usecols=(0,2,3), unpack='true')
sx3, sy3, serr3 = np.loadtxt("data/phase/solid_output_etot.dat", usecols=(0,2,3), unpack='true')
sx4, sy4, serr4 = np.loadtxt("data/phase/solid_output_temp.dat", usecols=(0,2,3), unpack='true')
sx5, sy5, serr5 = np.loadtxt("data/phase/solid_output_press.dat", usecols=(0,2,3), unpack='true')
```

Number of blocks = 100  
Number of steps per block = 1000

```python
fig = plt.figure(figsize=(20, 15))
#kin energy 
ax1 = fig.add_subplot(321)
ax1.errorbar(sx1, sy1, yerr=serr1)
ax1.set(xlabel="#blocks", ylabel="$E_\mathrm{kin}$/$N$")
ax1.set_xlim(1, 100)
ax1.grid(True)
#pot energy 
ax2 = fig.add_subplot(322)
ax2.errorbar(sx2, sy2, yerr=serr2)
ax2.set(xlabel="#blocks", ylabel="$E_\mathrm{pot}$/$N$")
ax2.set_xlim(1, 100)
ax2.grid(True)
#tot energy 
ax3 = fig.add_subplot(323)
ax3.errorbar(sx3, sy3, yerr=serr3)
ax3.set(xlabel="#blocks", ylabel="$E_\mathrm{tot}$/$N$")
ax3.set_xlim(1, 100)
ax3.grid(True)
#temp
ax4 = fig.add_subplot(324)
ax4.errorbar(sx4, sy4, yerr=serr4)
ax4.set(xlabel="#blocks", ylabel="$T$")
ax4.set_xlim(1, 100)
ax4.grid(True)
#pressure
ax5 = fig.add_subplot(325)
ax5.errorbar(sx5, sy5, yerr=serr5)
ax5.set(xlabel="#blocks", ylabel="$P$")
ax5.set_xlim(1, 100)
ax5.grid(True)
plt.show()
```
