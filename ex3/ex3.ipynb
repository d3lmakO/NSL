```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import math
from numpy import exp
from numpy.lib.scimath import log, sqrt
from scipy.special import erf
```

```python
def error(AV,AV2,n):
    if n==0:
        return 0
    else:
        return math.sqrt((AV2 - AV*AV)/n)
```

```python
def blocks(data_file):
    sum1 = 0.
    sum2 = 0.
    y_val = np.zeros((len(data_file), 2))
    for ix in range(len(data_file)):
        sum1 += data_file[ix]
        sum2 += data_file[ix] * data_file[ix]
        AV = sum1 / (ix + 1)
        AV2 = sum2 / (ix + 1)
        y_val[ix,0] = AV
        y_val[ix,1] = error(AV,AV2,ix)
    return y_val
```

## EX 3.1

We simulate Plain vanilla option pricing following Black-Scholes theory.  
Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits  
geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.



**For comparison** we calculate the expected price of **European Call-option price** and of **European Put-option price** using the analytical solution

Black-Scholes <span style="color:blue"> analytic solution </span> for the price at time t<T:
- European Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- European Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$,
and where
$${\rm erf}(x) = \frac{2}{\sqrt{\pi}}\int_{0}^{x} dz \, e^{-z^2}$$

```python
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0
#is the cumulative of Gaussian with mean = 0 and variance = 1
def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
d2 = d1 - sigma * sqrt(T)
call_opt = S0 * N(d1) - K * exp(-r * T) * N(d2)
put_opt = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
```

```python
with open("data/N_data_3_1.txt") as file:
    N_blocks = int(file.readline())
    N_throws = int(file.readline())
    N_steps = int(file.readline())
print(f"N_blocks = {N_blocks}")
print(f"N_throws = {N_throws}")
print(f"N_steps = {N_steps}")
#data
eu_call_direct = np.loadtxt("data/3_1_call_directly.txt")
eu_call_discrete = np.loadtxt("data/3_1_call_discretized.txt")
eu_put_direct = np.loadtxt("data/3_1_put_directly.txt")
eu_put_discrete = np.loadtxt("data/3_1_put_discretized.txt")
```

### 3.1.1 European Call-option price

1 - We compute at time $t=0$ **via Monte Carlo** the European call-option price  
By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$

```python
x_c = np.arange(N_blocks) * N_throws
y_c = blocks(eu_call_direct)
plt.errorbar(x_c, y_c[:,0] - call_opt, yerr=y_c[:,1])
plt.xlim(0, 1e6)
plt.xlabel(r"#throws")
plt.ylabel(r"$C(S(0), 0)$")
plt.grid(True)
plt.show()
print(f"C(S(0), 0) = {y_c[-1, 0] - call_opt:.4f} ± {y_c[-1, 1]:.4f}")
```

The result agrees with the analytical price.

2 - We estimate the European Call-option price using a discretize GBM

```python
x_c_d = np.arange(N_blocks) * N_throws
y_c_d = blocks(eu_call_discrete)
plt.errorbar(x_c_d, y_c_d[:,0] - call_opt, yerr=y_c_d[:,1])
plt.xlim(0, 1e6)
plt.xlabel(r"#throws")
plt.ylabel(r"$C(S(0), 0)$")
plt.grid(True)
plt.show()
print(f"C(S(0), 0) = {y_c_d[-1, 0] - call_opt:.4f} ± {y_c_d[-1, 1]:.4f}")
```

The result agrees with the analytical price.

### 3.1.2 European Put-option price

1 - European Put-option price directly sampled

```python
x_p = np.arange(N_blocks) * N_throws
y_p = blocks(eu_put_direct)
plt.errorbar(x_p, y_p[:,0] - put_opt, yerr=y_p[:,1])
plt.xlim(0, 1e6)
plt.xlabel(r"#throws")
plt.ylabel(r"$P(S(0), 0)$")
plt.grid(True)
plt.show()
print(f"P(S(0), 0) = {y_p[-1, 0] - put_opt:.4f} ± {y_p[-1, 1]:.4f}")
```

The result agrees with the analytical price.

2 - We estimate the European Put-option price using a discretize GBM

```python
x_p_d = np.arange(N_blocks) * N_throws
y_p_d = blocks(eu_put_discrete)
plt.errorbar(x_p_d, y_p_d[:,0] - put_opt, yerr=y_p_d[:,1])
plt.xlim(0, 1e6)
plt.xlabel(r"#throws")
plt.ylabel(r"$P(S(0), 0)$")
plt.grid(True)
plt.show()
print(f"P(S(0), 0) = {y_p_d[-1, 0] - put_opt:.4f} ± {y_p_d[-1, 1]:.4f}")
```
